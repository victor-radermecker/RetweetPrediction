```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../../data/Fuel.csv.csv')
```

```python
data = data[['ENGINESIZE', 'CO2EMISSIONS']]
data.head()
```

```python
train = data.iloc[:int(len(data)*0.8)]
test = data.iloc[int(len(data)*0.8):]
```

```python
X_train = np.array(train[['ENGINESIZE']])
y_train = np.array(train[['CO2EMISSIONS']])

reg = LinearRegression().fit(X_train, y_train)
```

```python
print("Score is ", reg.score(X_train, y_train))
print("Coefficient is: ", reg.coef_)
print("Intercept is: ", reg.intercept_)
```

### Metrics

```python
from sklearn.metrics import r2_score
```

```python
x_test = np.array(test[['ENGINESIZE']])
y_test = np.array(test[['CO2EMISSIONS']])
```

```python
predictions = reg.predict(x_test)

MAE = np.mean(np.absolute(y_test - predictions)) 
MSE = np.mean((y_test - predictions)**2)
r2_score = r2_score(y_test, predictions)
```

```python
print('MAE is', MAE)
print('MSE is', MSE)
print('R2 Score is ', r2_score)
```
